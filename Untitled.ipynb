```python
from sklearn import metrics
import numpy as np
from keras.datasets import fashion_mnist
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)

x_train = np.array(x_train).reshape(-1, 784)
x_test = np.array(x_test).reshape(-1, 784)
print(x_train.shape)
print(x_test.shape)
```

```python
from sklearn.externals import joblib
from sklearn import metrics
recogniser = joblib.load('fashion-mnist-model.pkl')
y_test_dt = recogniser.predict(x_test)
metrics.accuracy_score(y_test, recogniser.predict(x_test))
```

```python
label_dict = {
 0: "T-shirt/top",
 1: "Trouser",
 2: "Pullover",
 3: "Dress",
 4: "Coat",
 5: "Sanda",
 6: "Shirt",
 7: "Sneaker",
 8: "Bag",
 9: "Ankle boot"
}
```

```python
import seaborn as sn
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import preprocessing # import for normalise data
def plotter(test, output):
    array = metrics.confusion_matrix(test, output)
    array = preprocessing.scale(array) # to normalise
    df_cm = pd.DataFrame(array, index = [label_dict[i] for i in label_dict],
                      columns = [label_dict[i] for i in label_dict])
    plt.figure(figsize = (10,8))
    sn.heatmap(df_cm, cmap="Greens", annot=True, annot_kws={"size": 8})
    plt.savefig('confusion_matrix.png', dpi=600)
    plt.show()
    
plotter(y_test, y_test_dt)
```

```python
import random
def randNum():
    return random.randint(0,1001)

def batch_predict(ax, f1, f2, fontsize=12):
    num = randNum()
    value = recogniser.predict([x_test[num].reshape(784)])
    ax.axis('off')
    ax.imshow(x_test[num].reshape(28,28), cmap='gray_r')
    ax.set_title("Actual: " + label_dict[y_test[num]] + "\nPrediction: " + label_dict[value[0]] , fontsize=fontsize)
    
fig, axes = plt.subplots(4,4, figsize=(8,8))
[batch_predict(axes[attX, attY], attX, attY) for attX in [0,1,2,3] for attY in [0,1,2,3] ]
fig.subplots_adjust(wspace=1.5)
plt.show()


```
